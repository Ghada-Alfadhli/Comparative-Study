## **Getting started**
</div>

The document you are reading is not a static web page, but an interactive environment called a **Colab notebook** that lets you write and execute code.

For example, here is a **code cell** with a short Python script that computes a value, stores it in a variable, and prints the result:

```
seconds_in_a_day = 24 * 60 * 60
seconds_in_a_day
```

To execute the code in the above cell, select it with a click and then either press the play button to the left of the code, or use the keyboard shortcut "Command/Ctrl+Enter". To edit the code, just click the cell and start editing.

Variables that you define in one cell can later be used in other cells:

```
seconds_in_a_week = 7 * seconds_in_a_day
seconds_in_a_week
```

## Data science
</div>

With Colab you can harness the full power of popular Python libraries to analyze and visualize data. The code cell below uses **numpy** to generate some random data, and uses **matplotlib** to visualize it. To edit the code, just click the cell and start editing.

```
import numpy as np  # Import the NumPy library for numerical operations
import pandas as pd  # Import the Pandas library for data manipulation
import matplotlib.pyplot as plt  # Import the Matplotlib library for data visualization
import seaborn as sns  # Import the Seaborn library for improved data visualization
from sklearn.datasets import load_breast_cancer  # Import the Breast Cancer dataset from scikit-learn
from sklearn.model_selection import train_test_split  # Import the train-test split function
from sklearn.metrics import classification_report  # Import the classification report function for model evaluation
from sklearn.preprocessing import PolynomialFeatures  # Import the PolynomialFeatures for feature engineering
from sklearn.linear_model import LinearRegression  # Import the Linear Regression model
from sklearn.naive_bayes import GaussianNB  # Import the Gaussian Naive Bayes model
from sklearn.neighbors import KNeighborsClassifier  # Import the K-Nearest Neighbors classifier
from sklearn.linear_model import LogisticRegression  # Import the Logistic Regression model
from sklearn.preprocessing import StandardScaler  # Import the StandardScaler for data scaling


# Q1: Loading and preparing the dataset
data = load_breast_cancer()  # Load the Breast Cancer dataset
df = pd.DataFrame(data.data, columns=data.feature_names)  # Create a DataFrame with the dataset features
df['Diagnosis'] = data.target  # Add the 'Diagnosis' column to the DataFrame, which represents the target variable
print(df.columns)  # Display the column names

# Q2: Calculate and display the means and standard deviations for each feature
Means = df.mean()  # Calculate the mean for each feature
Stand_devs = df.std()  # Calculate the standard deviation for each feature
print("Means:")  # Display the means
print(Means)
print("Standard Deviations:")  # Display the standard deviations
print(Stand_devs)

# Q3: Determine the distribution of each feature (Normal or Non-Normal)
distributions = df.apply(lambda x: 'Normal' if np.abs(x.skew()) < 0.5 else 'Non-Normal')  # Check skewness for each feature
print(distributions)
# + With histogram diagram
# Define a function for plotting histograms of selected features from a data frame.
# The function takes the data frame, a list of feature names, and some plotting parameters as input.

def plot_feature_histograms(data_frame, feature_names, rows=6, cols=5, figsize=(15, 20)):
    # Create a new figure for plotting with the specified figure size.
    plt.figure(figsize=figsize)

    # Iterate through the feature names and plot histograms for each feature.
    for i, feature in enumerate(feature_names):
        # Create subplots in a grid format with the specified number of rows and columns.
        plt.subplot(rows, cols, i + 1)

        # Use Seaborn to create a histogram (histplot) for the current feature with a KDE overlay.
        sns.histplot(data_frame[feature], kde=True)

        # Add vertical lines for the mean and median values of the current feature.
        mean_value = data_frame[feature].mean()
        median_value = data_frame[feature].median()
        plt.axvline(mean_value, color='red', linestyle='dashed', linewidth=2, label='Mean')
        plt.axvline(median_value, color='green', linestyle='dashed', linewidth=2, label='Median')

        # Set the title for the subplot to the current feature's name and display the legend.
        plt.title(feature)
        plt.legend()

    # Adjust the layout to ensure proper spacing between subplots.
    plt.tight_layout()

# Usage
# Call the plot_feature_histograms function to generate histograms for the features in the 'bc' data frame
# using the feature names from 'data.feature_names'.
plot_feature_histograms(df, data.feature_names)



# Q4: Function to drop low-variance features
def drop_low_variance_features(df, threshold=0.01):  # Define a function to drop features with low variance
    low_variance_features = df.columns[df.var() < threshold]  # Find features with variance below the threshold
    df.drop(low_variance_features, axis=1, inplace=True)  # Drop those low-variance features from the DataFrame
drop_low_variance_features(df)  # Call the function to drop low-variance features
print(df.columns)  # Display the column names after dropping


# Q5: Count the number of examples for each prediction
prediction_counts = df['Diagnosis'].value_counts()  # Count the number of examples for each prediction class
print(prediction_counts)  # Display the counts

# Q6: Split the data into training and testing sets
x = df.drop('Diagnosis', axis=1)  # Features (X)
y = df['Diagnosis']  # Target variable (y)
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=80)  # Split the data into training and testing sets
print("X_train Shape:", X_train.shape)  # Display the shape of the training features
print("X_test Shape:", X_test.shape)  # Display the shape of the testing features
print("y_train Shape:", y_train.shape)  # Display the shape of the training target variable
print("y_test Shape:", y_test.shape)  # Display the shape of the testing target variable


## Q7 & Q8: Gaussian Naive Bayes and Polynomial Regression with Logistic Regression
# Gaussian Naive Bayes
gnb = GaussianNB()  # Create a Gaussian Naive Bayes model
gnb.fit(X_train, y_train)  # Train the model using the training data
gnb_predictions = gnb.predict(X_test)  # Predict using the test data
gnb_report = classification_report(y_test, gnb_predictions, target_names=data.target_names)  # Generate a classification report
print("Gaussian Naive Bayes Classification Report:")  # Display the classification report
print(gnb_report)

# Polynomial Features with Logistic Regression (for classification)
poly = PolynomialFeatures(degree=2)  # Create polynomial features with degree 2
X_train_poly = poly.fit_transform(X_train)  # Transform the training data into polynomial features
X_test_poly = poly.transform(X_test)  # Transform the test data into polynomial features

log_reg = LogisticRegression()  # Create a Logistic Regression model
log_reg.fit(X_train_poly, y_train)  # Train the model with polynomial features
poly_predictions = log_reg.predict(X_test_poly)  # Predict using the test data
poly_report = classification_report(y_test, poly_predictions, target_names=data.target_names)  # Generate a classification report
print("Polynomial Regression Classification Report:")  # Display the classification report
print(poly_report)


# K-Nearest Neighbors
knn = KNeighborsClassifier(n_neighbors=5)  # Initialize the K-Nearest Neighbors model with 5 neighbors
knn.fit(X_train, y_train)  # Train the model using the training data
knn_predictions = knn.predict(X_test)  # Predict using the test data
knn_report = classification_report(y_test, knn_predictions, target_names=data.target_names)  # Generate a classification report
print("K-Nearest Neighbors Classification Report:")  # Display the classification report
print(knn_report)
```
